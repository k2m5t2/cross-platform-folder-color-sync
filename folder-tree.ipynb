```python
GDRIVE = "/Users/yunhocho/Library/CloudStorage/GoogleDrive-yhjc2692@gmail.com/My Drive"
```

```python
import os

def scan_folder_tree(folder_path):
    folder_tree = {}
    for root, dirs, files in os.walk(folder_path):
        folder_name = os.path.basename(root)
        subfolders = [os.path.basename(subfolder) for subfolder in dirs]
        folder_tree[folder_name] = subfolders
    return folder_tree

# Specify the folder path you want to start scanning from
start_folder = f'{GDRIVE}'

# Call the function to scan the folder tree
folder_tree = scan_folder_tree(start_folder)

# Print the folder tree
print(folder_tree)
```
